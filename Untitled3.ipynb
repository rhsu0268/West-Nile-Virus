```python
def process_temperature(temp):
    temp = int(temp)
    if temp < 50:
        return 0.1
    elif temp < 60:
        return 0.3
    elif temp < 70:
        return 0.5
    elif temp < 80:
        return 0.7
    else:
        return 0.9
```

```python
joined_data_list = []
current_tuple = ()
for index, row in joined_data.iterrows():
    #print row
    temp = process_temperature(row['Tavg'])
    print row['Tavg']
    current_tuple = (row['Latitude'], row['Longitude'], temp)
    joined_data_list.append(current_tuple)
    #current_tuple = ()
```
